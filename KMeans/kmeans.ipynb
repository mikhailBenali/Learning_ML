```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
plt.scatter(X['petal length (cm)'], X['petal width (cm)'], c=iris.target)
```

```python
plt.scatter(X["sepal length (cm)"], X["sepal width (cm)"], c=iris.target)
```

```python
X = X.drop(['sepal length (cm)', 'sepal width (cm)'], axis='columns')
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
```

The best number of cluster is 2

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
plt.scatter(X['petal length (cm)'], X['petal width (cm)'], c=kmeans.labels_)
```
